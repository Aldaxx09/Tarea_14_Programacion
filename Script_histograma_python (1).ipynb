```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
os.chdir(r'C:\Users\alda09\MCD\SEMESTRE_1\CLASE_PROGRAMACION\Progarmacion_1')
import sys
```

```python


# Importamos el archivo
atletas = pd.read_csv("categorias_de_corredores.csv", sep=',', index_col=0)
# index_col=0 indicamos que las filas tienen un nombre

#Mostramos informacion mas detallada
atletas.info()

# Vemos las primeras filas del archivo
atletas.head()

```

```python


# Creamos el histograma de la variable tiempo
plt.figure(1)
plt.hist(atletas["Tiempo"], 15, color="red", ec="black")
plt.title("Histograma de Tiempo")
plt.savefig("Histograma.jpg")

# cambie el color del grafico a rojo
```

```python

# Para conocer la frecuencia de una variable que es categórica
import seaborn as sns

plt.figure(2)
sns.countplot(x=atletas['Velocidad'], palette='ocean')
plt.savefig('Velocidades.jpg')

```

```python
# Si queremos saber las velocidades en hombres y en mujeres
plt.figure(3)
grafico3 = sns.countplot(x='Genero', hue='Velocidad', palette='hot_r', data=atletas)
grafico3.set(title='Velocidades por Género',
             xlabel='Género', ylabel='Total')
plt.title("Velocidad por genero")
plt.savefig("Genero.jpg")
# plt.show()

# cambie el nombre del grafico
```
